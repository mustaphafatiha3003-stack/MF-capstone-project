```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# What is the churn distribution among patients?
churn_count = df['Churned'].value_counts()
churn_rate = df['Churned'].mean() * 100

print('CHURN DISTRIBUTION:')
print(churn_count)
print(f'\nOverall Churn Rate: {churn_rate:.2f}%')
print(f'- Churned Patients: {churn_count[1]} ({churn_count[1]/len(df)*100:.2f}%)')
print(f'- Retained Patients: {churn_count[0]} ({churn_count[0]/len(df)*100:.2f}%)')
```

```python
# Does churn increase with shorter or longer patient's tenure?
df['Tenure_group'] = pd.cut(df['Tenure_Months'],
    bins=[0, 12, 36, 60, 120],
    labels=['<1 year', '1–3 years', '3–5 years', '5+ years'])
tenure_churn = df.groupby('Tenure_group', observed= True)['Churned'].mean().reset_index()
tenure_churn['churn_rate_percent'] = tenure_churn['Churned'] * 100

print(tenure_churn)
```

```python
df.groupby('Churned')['Tenure_Months'].mean()
```

```python
# Does churn vary by age?
df['Age_group'] = pd.cut(df['Age'],
    bins=[0, 30, 45, 60, 100],
    labels=['18–30', '31–45', '46–60', '60+'])
age_churn = df.groupby('Age_group', observed= True)['Churned'].mean().reset_index()
age_churn['churn_rate_percent'] = age_churn['Churned'] * 100
print(age_churn)
```

```python
# Is there a difference in churn between male and female patients?
gender_churn = (df.groupby('Gender', observed=True).agg(
        patient_count=('Churned', 'count'),
        churn_rate=('Churned', 'mean')).reset_index())
gender_churn['churn_rate_percent'] = gender_churn['churn_rate'] * 100
print(gender_churn)
```

```python
df['State'].value_counts()
```

```python
df.groupby('State')['Churned'].mean().sort_values(ascending=False)
```

```python
# Does location influence patient churn?
location_churn = (df.groupby('State', observed=True).agg(
        patient_count=('Churned', 'count'),
        churn_rate=('Churned', 'mean')).reset_index())
location_churn['churn_rate_percent'] = location_churn['churn_rate'] * 100
print(location_churn)
```

```python
# Does distance to facility affect patient churn?
df.groupby('Churned')['Distance_To_Facility_Miles'].mean()
```

```python
df['Distance_group'] = pd.cut(df['Distance_To_Facility_Miles'],
    bins=[0, 5, 15, 30, 100],
    labels=['Very Close (0–5)', 'Close (6–15)', 'Far (16–30)', 'Very Far (30+)'])
distance_summary = (df.groupby('Distance_group', observed=True)
    .agg(patient_count=('Churned', 'count'),
        churn_rate=('Churned', 'mean')).reset_index())
distance_summary['churn_rate_percent'] = distance_summary['churn_rate'] * 100
print(distance_summary)
```

```python
# What are the key drivers of patient churn?
# The first; Service utilization
print('SERVICE UTILZATION ANALYSIS')
# Churn by medical specilty
specialty_churn = (df.groupby('Specialty', observed=True).agg(
        patient_count=('Churned', 'count'),
        churn_rate=('Churned', 'mean')).reset_index())

specialty_churn['churn_rate_percent'] = specialty_churn['churn_rate'] * 100
specialty_churn.sort_values('churn_rate_percent', ascending=False)
```

```python
# Churn by insurance type
insurance_churn = (df.groupby('Insurance_Type', observed=True)
    .agg(patient_count=('Churned', 'count'),
        churn_rate=('Churned', 'mean')).reset_index())

insurance_churn['churn_rate_percent'] = insurance_churn['churn_rate'] * 100
insurance_churn.sort_values('churn_rate_percent', ascending=False)
```

```python
# Visit frequency impact on churn
df['Visit_frequency'] = pd.cut(df['Visits_Last_Year'],
    bins=[0, 5, 10, 15],
    labels=['Infrequent (0–5)', 'Regular (6–10)', 'Frequent(10+)'])
visit_churn = (df.groupby('Visit_frequency', observed=True)
    .agg(patient_count=('Churned', 'count'),
        churn_rate=('Churned', 'mean')).reset_index())

visit_churn['churn_rate_percent'] = visit_churn['churn_rate'] * 100
visit_churn
```

```python
# Missed appointments impact on churn
df['Missed_group'] = pd.cut(df['Missed_Appointments'],
    bins=[-1, 0, 2, 5, 20],
    labels=['None', 'Low (1–2)', 'Medium (3–5)', 'High (6+)'])
missed_churn = (df.groupby('Missed_group', observed=True)
    .agg(patient_count=('Churned', 'count'),
        churn_rate=('Churned', 'mean')).reset_index())

missed_churn['churn_rate_percent'] = missed_churn['churn_rate'] * 100
missed_churn
```

```python
# Next is; Satisfaction score analysis
print('SATISFACTION SCORE ANALYSIS')
# Average satisfaction metrics by churn status
satisfaction_summary = (df.groupby('Churned').agg(
        avg_overall_satisfaction=('Overall_Satisfaction', 'mean'),
        avg_wait_time_satisfaction=('Wait_Time_Satisfaction', 'mean'),
        avg_staff_satisfaction=('Staff_Satisfaction', 'mean'),
        avg_provider_rating=('Provider_Rating', 'mean')).reset_index())
satisfaction_summary
```

```python
# Next is; Engagement and behavioural analysis
print('ENGAGEMENT AND BEHAVIOURAL ANALYSIS')
# Average days since last visit by churn status
# Do patients with longer time since last visit churn more?
df['Days_since_summary'] = pd.cut(df['Days_Since_Last_Visit'],
    bins=[0, 90, 180, 365, df['Days_Since_Last_Visit'].max()],
    labels=['Recent (<= 90 days)', 'Moderate (91-180)', 'Long(181-365)', 'Very long(365+)'],
    include_lowest=True)
days_since_summary_churn = (df.groupby('Days_since_summary', observed=True)
    .agg(patient_count=('Churned', 'count'),
        churn_rate=('Churned', 'mean')).reset_index())

days_since_summary_churn['churn_rate_percent'] = days_since_summary_churn['churn_rate'] * 100
days_since_summary_churn
```

```python
# Churn rate by portal usage
# Are digitally engaged patients less likely to churn?
portal_churn = (df.groupby('Portal_Usage').agg(
        patient_count=('Churned', 'count'),
        churn_rate=('Churned', 'mean')).reset_index())
portal_churn['churn_rate_percent'] = portal_churn['churn_rate'] * 100
portal_churn
```

```python
# Churn by referral behaviour
# Do patients who refer others stay longer?
referral_churn = (df.groupby('Referrals_Made').agg(
        patient_count=('Churned', 'count'),
        churn_rate=('Churned', 'mean')).reset_index())
referral_churn['churn_rate_percent'] = referral_churn['churn_rate'] * 100
referral_churn
```

```python
# Next is; Financial analysis
print('FINANCIAL ANALYSIS')
# Average out-of-pocket cost vs churn
# Do patients with higher personal healthcare costs churn more?
cost_summary = (df.groupby('Churned').agg(
        avg_out_of_pocket_cost=('Avg_Out_Of_Pocket_Cost', 'mean'),
        medium_out_of_pocket_cost=('Avg_Out_Of_Pocket_Cost', 'median'),
        high_out_of_pocket_cost = ('Avg_Out_Of_Pocket_Cost', 'max')).reset_index())
cost_summary
```

```python
df['Cost_summary'] = pd.cut(df['Avg_Out_Of_Pocket_Cost'],
    bins=[0, 500, 1000, 2000],
    labels=['Low (<= 500)', 'Medium (501-1000)', 'High(1000+)'])
cost_summary = (df.groupby('Cost_summary', observed=True)
    .agg(patient_count=('Churned', 'count'),
        churn_rate=('Churned', 'mean')).reset_index())

cost_summary['churn_rate_percent'] = cost_summary['churn_rate'] * 100
cost_summary
```

```python
# Does experiencing billing issues increase patient churn?
billing_churn = (df.groupby('Billing_Issues')
    .agg(patient_count=('Churned', 'count'),
        churn_rate=('Churned', 'mean')).reset_index())
billing_churn['churn_rate_percent'] = billing_churn['churn_rate'] * 100
billing_churn
```

```python
# Calculation of revenue at risk
# Revenue at risk from churned patients
df['Revenue_at_risk'] = df['Avg_Out_Of_Pocket_Cost'] * df['Churned']
total_revenue_at_risk = df['Revenue_at_risk'].sum()
print(f'\nTOTAL REVENUE AT RISK: {total_revenue_at_risk:.2f}')
```

```python
# Finally; Risk profiling
print('RISK PROFILING')
# which patients segment are at highest risk of churning (risk profiling)
df['Low_satisfaction_flag'] = df['Overall_Satisfaction'] <= 2
df['Long_inactive_flag'] = df['Days_Since_Last_Visit'] > 90
df['High_cost_flag'] = df['Avg_Out_Of_Pocket_Cost'] > df['Avg_Out_Of_Pocket_Cost'].median()
df['Billing_issue_flag'] = df['Billing_Issues'] == 1
df['Low_engagement_flag'] = ((df['Portal_Usage'] == 0) & (df['Referrals_Made'] == 0))

risk_columns = ['Low_satisfaction_flag','Long_inactive_flag','High_cost_flag','Billing_issue_flag',
    'Low_engagement_flag']
df['churn_risk_score'] = df[risk_columns].sum(axis=1)
df['Churn_risk_level'] = pd.cut(df['churn_risk_score'],
     bins=[-1, 1, 3, 5],
     labels=['Low Risk', 'Medium Risk', 'High Risk'])
risk_validation = (df.groupby('Churn_risk_level', observed=True).agg(
        patient_count=('Churned', 'count'),
        churn_rate=('Churned', 'mean')).reset_index())
risk_validation['churn_rate_percent'] = risk_validation['churn_rate'] * 100
risk_validation
```

```python
# Highest risk patient segment
high_risk_profile = (df[df['Churn_risk_level'] == 'High Risk']
    .groupby('Churned').agg(
        avg_age=('Age', 'mean'),
        avg_tenure=('Tenure_Months', 'mean'),
        avg_days_since_last_visit=('Days_Since_Last_Visit', 'mean'),
        avg_cost=('Avg_Out_Of_Pocket_Cost', 'mean'),
        billing_issue_rate=('Billing_Issues', 'mean'),
        portal_usage_rate=('Portal_Usage', 'mean'),
        referral_rate=('Referrals_Made', 'mean')))
high_risk_profile
```
